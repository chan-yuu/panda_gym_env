## Interract

```python
import gymnasium as gym
import panda_gym
from numpngw import write_apng
from IPython.display import Image

env = gym.make("PandaPickAndPlace-v3", render_mode="rgb_array")
env = gym.make("PandaMobilePickAndPlace-v3", render_mode="rgb_array")
observation, info = env.reset()

images = [env.render()]
for _ in range(3):
    print(_)
    action = env.action_space.sample()
    observation, reward, terminated, truncated, info = env.step(action)
    images.append(env.render())

    if terminated or truncated:
        observation, info = env.reset()
        images.append(env.render())

env.close()
```

## Convert into animation


```python
write_apng("anim.gif", images, delay=40)  # real-time rendering = 40 ms between frames
```
